```python
from numpy import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Limites de l'intégrale
a = 0
b = np.pi
N = 1000

def f(x):
  return (np.sin(x)**3 + np.cos(x)**5) / np.sqrt(1 + x **4)

#Pour stocker les valeurs
vals = []

for i in range(N) :
  ar = np.zeros(N)

  for i in range(len(ar)):
    ar[i] = random.uniform(a, b)

  integral1 = 0.0
  for i in ar :
    integral1 += f(i)

  ans = (b-a)/float(N) * integral1
  vals.append(ans)

print("Value calculated by a monte carlo integration {}.".format(ans))
plt.title("Distributions of areas calculated")
plt.hist(vals, bins=30, ec="black")
plt.xlabel("Areas")
plt.show()
```
